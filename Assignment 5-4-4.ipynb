JSON example, with string
demonstrates creation of normalized dataframes (tables) from nested json string
source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

JSON exercise
Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above, 1. Find the 10 countries with most projects 2. Find the top 10 major project themes (using column 'mjtheme_namecode') 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Problem 1: Find the countries with the most projects
data=pd.read_json('world_bank_projects.json') #read in data file
data.countryname.value_counts().sort_values(ascending=False).head(10)
```

```python
#Problem 3: Fill in missing names based on the codes
#Doing this first so I can do problem 2 properly, as I can't take an accurate count of names with missing values
name_list=sum(data.mjtheme_namecode.to_list(),[]) #turn series of dicts into flattened list of dicts
df_themes=json_normalize(name_list) #turn flattened list into its own dataframe
df_themes.head(20) #see missing values
```

```python
df=df_themes
dummy=df.sort_values(by=['code','name']) #Sorts so that all "null" values are at the top of each section of the code column
filled_df=dummy.replace(to_replace='',value=None, method='bfill') #back fill missing values with correct values (works because of way dataframe is sorted)
filled_df.sort_index().head(20) #sort dataframe back it its original state
```

```python
#Problem 2:find the top ten project themes
filled_df.name.value_counts().head(10)
```
